# Imports used

```python
import os
import librosa
import librosa.display
import numpy as np
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,precision_score, recall_score, f1_score


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical

from keras.preprocessing import image
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Flatten, Dense
```

# MFCC for KNN and RNN

```python
# Define feature extraction function
def extract_mfcc_features(file_path, n_mfcc=13):
    y, sr = librosa.load(file_path)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfccs.T, axis=0)
```

```python
# Define data loading function
def load_mfcc_data(data_dir, n_classes=9, n_mfcc=13):
    features = []
    labels = []
    for label in range(n_classes):
        class_dir = os.path.join(data_dir, str(label))
        for file_name in os.listdir(class_dir):
            file_path = os.path.join(class_dir, file_name)
            mfccs = extract_mfcc_features(file_path, n_mfcc)
            features.append(mfccs)
            labels.append(label)
    return np.array(features), np.array(labels)
```

```python
# Load dataset and extract features
data_dir = 'Datasets/Augment_KNN_RNN/'  # Path to your dataset directory
n_classes = 10  # Number of classes (9 classes of gunshots & 1 classes of non-gunshots)
n_mfcc = 13  # Number of MFCC features

mfcc_features, labels = load_mfcc_data(data_dir, n_classes, n_mfcc)
```

# Train and Evaluation KNN Model

```python
# Standardize MFCC data for KNN
scaler = StandardScaler()
mfcc_features_standardized = scaler.fit_transform(mfcc_features)
```

```python
# Split data into training and testing sets
X_train_knn, X_test_knn, y_train_knn, y_test_knn = train_test_split(mfcc_features_standardized, labels, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(X_train_knn, y_train_knn)
y_pred_knn = knn.predict(X_test_knn)
```

```python
# KNN Model Evaluation Metrics
accuracy_knn = accuracy_score(y_test_knn, y_pred_knn)
precision_knn = precision_score(y_test_knn, y_pred_knn, average='weighted')
recall_knn = recall_score(y_test_knn, y_pred_knn, average='weighted')
f1_knn = f1_score(y_test_knn, y_pred_knn, average='weighted')
conf_matrix_knn = confusion_matrix(y_test_knn, y_pred_knn)

print("KNN Model Classification Report:")
print(classification_report(y_test_knn, y_pred_knn))

print(f'KNN Model - Accuracy: {accuracy_knn}')
print(f'KNN Model - Precision: {precision_knn}')
print(f'KNN Model - Recall: {recall_knn}')
print(f'KNN Model - F1 Score: {f1_knn}')
```

```python
# Plot confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_knn, annot=True, fmt='d', cmap='Blues')
plt.title('KNN Model Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

# Train and Evaluation RNN Model

```python
# Reshape Features for RNN input
mfcc_features_rnn = mfcc_features.reshape(mfcc_features.shape[0], mfcc_features.shape[1], 1)

# Convert labels to categorical
labels_rnn = to_categorical(labels, num_classes=n_classes)

# Split the data into training and testing sets
X_train_rnn, X_test_rnn, y_train_rnn, y_test_rnn = train_test_split(mfcc_features_rnn, labels_rnn, test_size=0.2, random_state=42)
```

```python
# RNN model
rnn_model = Sequential()
rnn_model.add(LSTM(128, input_shape=(n_mfcc, 1), return_sequences=True))
rnn_model.add(Dropout(0.2))
rnn_model.add(LSTM(64, return_sequences=False))
rnn_model.add(Dropout(0.2))
rnn_model.add(Dense(32, activation='relu'))
rnn_model.add(Dense(n_classes, activation='softmax'))
rnn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the RNN model and store the history
history = rnn_model.fit(X_train_rnn, y_train_rnn, epochs=50, batch_size=32, validation_data=(X_test_rnn, y_test_rnn))
```

```python
y_pred_rnn = np.argmax(rnn_model.predict(X_test_rnn), axis=1)
y_test_class_rnn = np.argmax(y_test_rnn, axis=1)
```

```python
# RNN Model Evaluation Metrics
accuracy_rnn = accuracy_score(y_test_class_rnn, y_pred_rnn)
precision_rnn = precision_score(y_test_class_rnn, y_pred_rnn, average='weighted')
recall_rnn = recall_score(y_test_class_rnn, y_pred_rnn, average='weighted')
f1_rnn = f1_score(y_test_class_rnn, y_pred_rnn, average='weighted')
conf_matrix_rnn = confusion_matrix(y_test_class_rnn, y_pred_rnn)

print("RNN Model Classification Report:")
print(classification_report(y_test_class_rnn, y_pred_rnn))

print(f'RNN Model - Accuracy: {accuracy_rnn}')
print(f'RNN Model - Precision: {precision_rnn}')
print(f'RNN Model - Recall: {recall_rnn}')
print(f'RNN Model - F1 Score: {f1_rnn}')
```

```python
# Plot confusion matrix for RNN
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_rnn, annot=True, fmt='d', cmap='Blues')
plt.title('RNN Model Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
# Plot accuracy curve for RNN
plt.figure(figsize=(10, 7))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('RNN Model Accuracy Curve')
plt.legend()
plt.show()
```

```python
def create_spectrogram(audio_file, image_file):
    # Adjust figure size if needed
    fig = plt.figure(figsize=(8, 6))  # You can adjust the figure size as needed
    ax = fig.add_subplot(1, 1, 1)
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

    # Load audio file
    y, sr = librosa.load(audio_file)

    # Compute mel spectrogram
    # Adjust parameters such as n_fft, hop_length, and n_mels for better resolution
    # You may also experiment with window type (e.g., window='hann') and other parameters
    ms = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512, n_mels=128)

    # Convert power to decibels (log scale)
    log_ms = librosa.power_to_db(ms, ref=np.max)

    # Display the mel spectrogram
    librosa.display.specshow(log_ms, sr=sr, x_axis='time', y_axis='mel')

    # Add color bar for intensity scale (optional)
    plt.colorbar(format='%+2.0f dB')

    # Save the spectrogram as an image file
    fig.savefig(image_file, bbox_inches='tight')  # Use bbox_inches='tight' to prevent clipping
    plt.close(fig)
```

```python
def create_pngs_from_wavs(input_path, output_path):
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    dir = os.listdir(input_path)

    for i, file in enumerate(dir):
        if not file.endswith('.wav'):
            continue  # Skip non-WAV files

        input_file = os.path.join(input_path, file)
        output_file = os.path.join(output_path, file.replace('.wav', '.png'))
        create_spectrogram(input_file, output_file)
```

```python
def load_images_from_path(path, label):
    images = []
    labels = []
    for file in os.listdir(path):
        # Ignore directories
        if os.path.isfile(os.path.join(path, file)):
            images.append(image.img_to_array(image.load_img(os.path.join(path, file), target_size=(224, 224, 3))))
            labels.append(label)
    return images, labels

def show_images(images):
    fig, axes = plt.subplots(1, 8, figsize=(20, 20), subplot_kw={'xticks': [], 'yticks': []})

    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i] / 255)
        
x = []
y = []
```

```python
classes = ['AK-12', 'AK-47', 'IMI Desert Eagle', 'M16', 'M249', 'M4', 'MG-42', 'MP5', 'Zastava M92',
                    'NonGunshots']
```

```python
for i, cls in enumerate(classes):
    images, labels = load_images_from_path(f'Spectrograms_New/{cls}', i)
    x += images
    y += labels
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.2, random_state=42)

x_train_norm = np.array(x_train) / 255
x_test_norm = np.array(x_test) / 255

y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)
```

```python
#CNN Model

cnn_model = Sequential()
cnn_model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
cnn_model.add(MaxPooling2D(2, 2))
cnn_model.add(Conv2D(128, (3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(2, 2))
cnn_model.add(Conv2D(128, (3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(2, 2))
cnn_model.add(Conv2D(128, (3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(2, 2))
cnn_model.add(Flatten())
cnn_model.add(Dense(1024, activation='relu'))
cnn_model.add(Dense(10, activation='softmax'))
cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn_model.summary()
```

```python
cnn_history = cnn_model.fit(x_train_norm, y_train_encoded, epochs=15, batch_size=32, validation_data=(x_test_norm, y_test_encoded))
```
